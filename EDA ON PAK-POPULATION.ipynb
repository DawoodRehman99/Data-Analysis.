**Dataset Overview:**
This dataset contains demographic data from the 2017 Pakistan Population Census. It provides detailed information on population distribution across provinces, divisions, districts, and sub-divisions. Key details include urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from 1998, and annual growth rates.

Key Features:

*Province: Administrative regions where data was collected.
Division: Second-level administrative units within provinces.
District: Larger administrative areas within divisions.
Sub-Division: Localized areas (tehsils) within districts.
Area: Land area of each sub-division (in square kilometers).
Urban Population 2017: Urban population per sub-division in 2017.
Rural Population 2017: Rural population per sub-division in 2017.
Male & Female Population 2017: Gender-specific population data.
Transgender Population 2017: Transgender population per sub-division.
Sex Ratio 2017: Number of females per 1000 males in each sub-division.
Population 1998: Population recorded in the 1998 census.
Annual Growth Rate: Population growth rate from 1998 to 2017.
Source: Official reports from the Pakistan Population Census.*

Usage: Researchers can use this data to analyze demographic trends, population growth, and gender distribution at different levels of Pakistanâ€™s administrative structure. Ensure proper attribution and compliance with legal guidelines.

### We Have Explored Numerical Insights; Now Let's Visualize Outliers and Distributions Through Plots.

```python
box_plots_per_figure = 4  


for i in range(0, len(pak_population_data.select_dtypes(include=['float64', 'int64']).columns), box_plots_per_figure):
    plt.figure(figsize=(15, 10))
    
 
    for j, column in enumerate(pak_population_data.select_dtypes(include=['float64', 'int64']).columns[i:i + box_plots_per_figure]):
        plt.subplot(2, 2, j + 1)
        pak_population_data.boxplot(column=column)
        
        mean_value = pak_population_data[column].mean()
        plt.axhline(mean_value, color='red', linestyle='--', label='Mean')
        
        plt.title(column)
        plt.ylabel('Values')
        plt.legend()
    
    plt.tight_layout()  
    plt.show()  
```

### Identify and Remove outliers

```python
sns.boxplot(x='PROVINCE', y='AREA (sq.km)', data=pak_population_data)
```

The data shows several outliers across provinces, with Balochistan having the most, reaching up to 17,500 sq. km. Balochistan also has a large interquartile range (IQR), indicating high variability in area sizes, while Sindh and Punjab show more consistency with smaller IQRs. KPK and KPK/FATA have the smallest IQRs, suggesting minimal variation in area sizes. The higher median area size in Balochistan suggests generally larger areas than other provinces, where medians indicate smaller, more consistent area sizes.

```python
# Assuming your DataFrame is named 'df' and the relevant columns are 'PROVINCE' and 'AREA'
q1 = df.groupby('PROVINCE')['AREA (sq.km)'].quantile(0.25)
q3 = df.groupby('PROVINCE')['AREA (sq.km)'].quantile(0.75)
print(q1)
print(q3)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(lower_bound)
print(upper_bound)
```

```python
# Merging lower and upper bounds back into the original DataFrame
df = df.merge(lower_bound, on='PROVINCE', suffixes=('', '_lower'))
df = df.merge(upper_bound, on='PROVINCE', suffixes=('', '_upper'))
```

```python
# Removing outliers
df_no_outliers = df[(df['AREA (sq.km)'] >= df['AREA (sq.km)_lower']) & (df['AREA (sq.km)'] <= df['AREA (sq.km)_upper'])]

# Dropping the temporary bound columns
df_no_outliers = df_no_outliers.drop(columns=['AREA (sq.km)_lower', 'AREA (sq.km)_upper'])
```

```python
print("Original DataFrame size:", df.shape)
print("DataFrame size after removing outliers:", df_no_outliers.shape)
```
